## Data Cleaning

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)


# Fill missing categorical columns with mode (most frequent value)
for col in data.select_dtypes(include=['object']).columns:
    data[col].fillna(data[col].mode()[0], inplace=True)

#numerical
for col in data.select_dtypes(include=[int, float]).columns:
    if data[col].notnull().any():
        data[col].fillna(data[col].mean(), inplace=True)
```

```python
#Recheck
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Ensure all columns have the correct data type
data.dtypes
```

```python
def treat_outliers_all_columns(df):
    for col in df.select_dtypes(include=['number']).columns:
        # Calculate the 95th percentile
        q95 = np.quantile(df[col], 0.95)

        # Identify outliers
        outliers = df[col] > q95

        # Replace outliers with the median
        df.loc[outliers, col] = df[col].median()

    return df

treat_outliers_all_columns(data)
```

```python
data['Total Data (Bytes)'] = data['Total DL (Bytes)'] + data['Total UL (Bytes)']
data['Total Duration (s)'] = data['Dur. (ms)'] / 1000  # am converting to secs

# Segment the users into deciles based on the total duration
data['Duration Decile'] = pd.qcut(data['Total Duration (s)'], q=5, labels=False) + 1
# Group by the decile class and compute the sum of total data usage for each decile
decile_data = data.groupby('Duration Decile')['Total Data (Bytes)'].sum().reset_index()

decile_data.columns = ['Decile Class', 'Total Data (Bytes)']
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(decile_data['Decile Class'], decile_data['Total Data (Bytes)'], color='skyblue')
plt.title('Total Data Usage by Duration Decile Class')
plt.xlabel('Duration Decile Class')
plt.ylabel('Total Data (Bytes)')
plt.xticks(decile_data['Decile Class'])
plt.grid(True)
plt.show()
```
